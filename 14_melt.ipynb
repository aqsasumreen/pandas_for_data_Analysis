```python
# melt functiion use to reshape dataframe
import pandas as pd
a = pd.DataFrame({"A": [1, 2, 3,4], "B": [5, 6, 7, 8], "C": [11, 12, 13,12]})
pd.melt(a)
# show the data frame vertically

```

```python
pd.melt(a, id_vars= ["B"],var_name = "python")
```

```python
# pivot

a = pd.DataFrame({"A": [1, 2, 3,4], "name": ["a", "b", "c", "d"], "C": [11, 12, 13,12], "D": [21, 22, 23, 24]})
a.pivot(index="A", columns = "name")
# A show as index and arrange according to names
```

```python
a = pd.DataFrame({"A": [1, 2, 3,4], "name": ["a", "b", "c", "d"], "C": [11, 12, 13,12], "D": [21, 22, 23, 24]})
a.pivot(index="A", columns = "name", values = "D")
# sirf D ka data mily ga
```

```python
a = pd.DataFrame({"A": [1, 2, 3,4], "name": ["a", "b", "c", "d"], "C": [11, 12, 13,12], "D": [21, 22, 23, 24]})
a.pivot_table(index="name", columns = "A", aggfunc="sum", margins=True)
```
